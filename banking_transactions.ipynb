```python
#Use this script to convert the parse_transactions.py csv output to a clean date.


import pandas as pd

#Import the output file of the parse_transactions.py program
bk = pd.read_csv("./ofxparse/output.csv")

#Convert the date to readable
bk["Date"] = pd.to_datetime(bk["Date"])

#Change date formatting to mm/dd/yyyy
bk["Date"] = bk["Date"].dt.strftime("%m/%d/%Y")
bk.head(10)

#Write transactions to csv without keeping the index.
bk.to_csv("./data/banking_transactions.csv", index=False)
```

```python
import pandas as pd
from time import sleep
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="myGeocoder")

with open("./data/addresses.txt", "r") as file:
    addresses = [line.strip() for line in file if line.strip()]

results = []

for addr in addresses:
    try:
        location = geolocator.geocode(addr)
        if location:
            coords = f"{location.latitude}, {location.longitude}"
        else:
            coords = None
    except Exception:
        coords = None

    results.append({"address": addr, "coordinates": coords})

    print(f"Address: {addr} -> Coordinates: {coords}")
    sleep(1)

#Save results to CSV
pd.DataFrame(results).to_csv("./data/address_coordinates.csv", index=False)
```
