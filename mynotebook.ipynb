## Steam Games loc&iloc

```python
import pandas as pd
sg = pd.read_csv('./data/steam_games_dataset.csv') #Steam Games dataset
```

```python
sg.loc[1, "price"] = 6767
```

```python
sg.loc[1:99,["name","positive","negative","price"]]
```

```python
sg.iloc[0:11, [1,5,6]]
```

```python
sg.loc[sg['price'] > 1000, ['name', 'positive', 'price']]
```

```python
sg.loc[(sg['price'] > 1000) & (sg['positive'] > 1000000), ['name', 'positive', 'price']]
```

```python
sg[sg['name'].str.startswith("Touhou")]
```

```python
sg.query('price == 599')
```

```python
sg_modified = sg.copy()
#sg.drop(columns=['score_rank'])
sg_modified.drop(columns=['score_rank','owners','average_forever','average_2weeks','userscore','median_forever','median_2weeks','discount','ccu'], inplace=True)
sg_modified
```

```python
#Convert prices to decimal.
sg_modified['price']= sg_modified['price'] / 100
sg_modified['initialprice'] = sg_modified['initialprice'] / 100
sg_modified
```

## Most and least expensive games

```python
# Top 10 most expensive games
sg_modified.nlargest(10, 'price')
```

```python
# Top 10 cheapest games that aren't 0.00
sg_modified[sg_modified['price'] > 0].nsmallest(10, 'price')
```
